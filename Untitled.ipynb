```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#Ratio of almost 1
#Should hover around 1

free = []
paid = []

with open(r"C:\\Users\\Charlie\\Desktop\\free_response_R1.txt", "r") as file:
    for readline in file: 
        line_strip = readline.strip()
        free.append(int(line_strip))
with open(r"C:\\Users\\Charlie\\Desktop\\paid_response_R1.txt", "r") as file:
    for readline in file: 
        line_strip = readline.strip()
        paid.append(int(line_strip))

if len(paid) > len(free):
    paid = paid[0:len(free)]
elif len(free) > len(paid):
    free = free[0:len(paid)]
    
ratio = []
for i in range(len(paid)):
    if free[i] == 0:
        free[i] = .000001
    elif paid[i] == 0:
        paid[i] = .00001
    ratio.append(free[i]/paid[i])

if len(paid) > len(free):
    paid = paid[0:len(free)]
elif len(free) > len(paid):
    free = free[0:len(paid)]
    
x = range(len(paid))

d = {'actual_ratio':ratio, 'time':x}

df = {'free':free, 'time':x}
dp = {'paid':paid, 'time':x}

sns.set_theme()
sns.set(rc={"figure.figsize":(10, 4)})

tips = pd.DataFrame(data=d)

sns.relplot(
    data=tips, x="time", y="actual_ratio", aspect=2)

plt.ylim(0,2)
plt.show()

tips = pd.DataFrame(data=df)

sns.relplot(
    data=tips, x="time", y="free", aspect=2)

tips = pd.DataFrame(data=dp)

sns.relplot(
    data=tips, x="time", y="paid", aspect=2)
```
